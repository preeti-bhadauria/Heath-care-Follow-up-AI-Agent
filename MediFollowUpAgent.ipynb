```python
#Task 2 : define agent tools

def get_patient(patient_id):
    """
    Retrieve complete patient information.
    """
    patient = df[df["patient_id"] == patient_id]

    if patient.empty:
        return None

    return patient.iloc[0].to_dict()
```

```python
get_patient("P0001")
```

```python
def get_vitals(patient_id):
    """
    Retrieve patient vital signs.
    """
    patient = get_patient(patient_id)

    if patient is None:
        return {}

    return {
        "Blood Pressure":
            f"{patient['vitals_bp_systolic']}/{patient['vitals_bp_diastolic']} mmHg",

        "Heart Rate":
            patient["vitals_heart_rate"],

        "SpO2":
            patient["vitals_spo2"]
    }
```

```python
get_vitals("P0001")
```

```python
def get_lab_results(patient_id):
    """
    Retrieve latest laboratory result.
    """
    patient = get_patient(patient_id)

    if patient is None:
        return {}

    return {
        "Lab Test":
            patient["lab_test"],

        "Lab Value":
            patient["lab_value"],

        "Unit":
            patient["lab_unit"]
    }
```

```python
get_lab_results("P0001")
```

```python
df["last_visit_date"] = pd.to_datetime(df["last_visit_date"])
df["next_scheduled_visit"] = pd.to_datetime(df["next_scheduled_visit"])
```

```python
def get_appointment(patient_id):
    """
    Retrieve appointment information.
    """
    patient = get_patient(patient_id)

    if patient is None:
        return {}

    return {
        "Last Visit":
            str(patient["last_visit_date"].date()),

        "Next Visit":
            str(patient["next_scheduled_visit"].date()),

        "Days Since Last Visit":
            patient["days_since_last_visit"],

        "Missed Appointment":
            patient["missed_last_appointment"]
    }
```

```python
get_appointment("P0001")
```

```python
def get_notes(patient_id):
    """
    Retrieve clinician notes.
    """
    patient = get_patient(patient_id)

    if patient is None:
        return ""

    return patient["notes"]
```

```python
tool_map = {
    "get_patient": get_patient,
    "get_vitals": get_vitals,
    "get_lab_results": get_lab_results,
    "get_notes": get_notes,
    "get_appointment": get_appointment
}
```

```python
tool_description = """
Available tools:

1. get_patient(patient_id)
Returns demographics and diagnosis.

2. get_vitals(patient_id)
Returns BP, Heart Rate, SpO2.

3. get_lab_results(patient_id)
Returns laboratory test results.

4. get_notes(patient_id)
Returns doctor's notes.

5. get_appointment(patient_id)
Returns appointment history.

When you need information,
ONLY respond with

TOOL:tool_name(patient_id)

Example

TOOL:get_patient(P0001)

Do not answer the question until all necessary tools have been used.

When enough information is available,
provide the final clinical assessment.
"""
```

```python
tools = [
    {
        "name": "get_patient",
        "description": "Retrieve complete patient demographics and diagnosis.",
        "parameters": {
            "type": "OBJECT",
            "properties": {
                "patient_id": {
                    "type": "STRING"
                }
            },
            "required": ["patient_id"]
        }
    },

    {
        "name": "get_vitals",
        "description": "Retrieve patient's vital signs.",
        "parameters": {
            "type": "OBJECT",
            "properties": {
                "patient_id": {
                    "type": "STRING"
                }
            },
            "required": ["patient_id"]
        }
    },
    
    {
        "name": "get_lab_results",
        "description": "Retrieve laboratory results.",
        "parameters": {
            "type": "OBJECT",
            "properties": {
                "patient_id": {
                    "type": "STRING"
                }
            },
            "required": ["patient_id"]
        }
    },

    {
        "name": "get_notes",
        "description": "Retrieve doctor's clinical notes.",
        "parameters": {
            "type": "OBJECT",
            "properties": {
                "patient_id": {
                    "type": "STRING"
                }
            },
            "required": ["patient_id"]
        }
    },

    {
        "name": "get_appointment",
        "description": "Retrieve appointment history.",
        "parameters": {
            "type": "OBJECT",
            "properties": {
                "patient_id": {
                    "type": "STRING"
                }
            },
            "required": ["patient_id"]
         }
    }
]           
```

```python
def patient_agent(patient_id):

    history = []

    prompt = f"""
You are an experienced clinical care coordinator.

Patient ID: {patient_id}

You have access to the following tools:

get_patient
get_vitals
get_lab_results
get_notes
get_appointment

If additional information is required:

- Call ONLY ONE tool at a time.
- Never return multiple TOOL lines.
- Never call the same tool twice.
- After receiving a tool result, decide whether another tool is needed.
- Once you have enough information, stop calling tools and generate the final report.

Respond with exactly one of the following:

TOOL:get_patient

OR

TOOL:get_vitals

OR

TOOL:get_lab_results

OR

TOOL:get_notes

OR

TOOL:get_appointment

Once enough information has been collected, provide:

1. Clinical Risk
2. Key Findings
3. Recommended Follow-up
4. Priority
"""

    while True:

        response = client.models.generate_content(
            model="gemini-3.1-flash-lite",
            contents=prompt + "\n\n" + str(history)
        )

        reply = response.text.strip()

        print("\nGemini:", reply)

        if not reply.startswith("TOOL:"):
            return reply

        tool_name = reply.replace("TOOL:", "").strip()

        if tool_name not in tool_map:
            return f"Unknown tool: {tool_name}"

        print(f"\nExecuting {tool_name}...\n")

        result = tool_map[tool_name](patient_id)

        history.append({
            "tool": tool_name,
            "result": result
        })
```

```python
def assign_priority(row):

    diagnosis = str(row["diagnosis"]).lower()

    bp = row["vitals_bp_systolic"]

    days = row["days_since_last_visit"]

    lab = row["lab_value"]

    # HIGH PRIORITY
    if (
        "heart failure" in diagnosis
        or "chronic kidney disease" in diagnosis
        or "ckd" in diagnosis
        or bp >= 160
        or lab > 9
        or days > 180
    ):
        return "High"

    # MEDIUM PRIORITY
    elif (
        "diabetes" in diagnosis
        or "hypertension" in diagnosis
        or "asthma" in diagnosis
        or "anemia" in diagnosis
        or "depression" in diagnosis
    ):
        return "Medium"

    else:
        return "Low"


df["followup_priority"] = df.apply(assign_priority, axis=1)
```

```python
#Gemini follow up agent


def followup_agent(patient_id):

    prompt = f"""
Patient {patient_id} missed their last appointment.

Use available patient information to recommend:

1. Follow-up Priority
2. Why follow-up is needed
3. Suggested Action
4. Timeframe
"""

    response = client.models.generate_content(
        model="gemini-3.1-flash-lite",
        contents=prompt
    )

    return response.text
```

```python
results = []

for pid in missed_df["patient_id"]:

    print("="*70)
    print(pid)

    recommendation = followup_agent(pid)

    print(recommendation)

    results.append(
        {
            "Patient ID": pid,
            "Recommendation": recommendation
        }
    )
```

```python
def followup_agent(patient_id):

    patient = get_patient(patient_id)
    appointment = get_appointment(patient_id)
    vitals = get_vitals(patient_id)
    labs = get_lab_results(patient_id)

    prompt = f"""
    Patient Information:
    {patient}

    Appointment:
    {appointment}

    Vitals:
    {vitals}

    Labs:
    {labs}

    The patient missed their last appointment.

    Recommend:
    1. Follow-up Priority
    2. Why follow-up is needed
    3. Suggested Action
    4. Timeframe
    """

    try:
        response = client.models.generate_content(
            model="gemini-3.1-flash-lite",
            contents=prompt
        )
        return response.text

    except Exception as e:
        return f"Error: {e}"
```
